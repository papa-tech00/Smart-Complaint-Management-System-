```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv("../data/complaints_clean.csv")

print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Input feature
X = df["clean_text"]

# Target label
y = df["category"]

print("Input (X):")
print(X.head())

print("\nTarget (y):")
print(y.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words="english"
)

# Convert text into numerical features
X_features = vectorizer.fit_transform(X)

# Print shape
print("TF-IDF Matrix Shape:", X_features.shape)
```

```python
print(vectorizer.get_feature_names_out()[:20])
```

```python
from sklearn.preprocessing import LabelEncoder

# Create Label Encoder
label_encoder = LabelEncoder()

# Convert categories into numbers
y_encoded = label_encoder.fit_transform(y)

# Print encoded labels
print("Encoded Labels:")
print(y_encoded[:10])
```

```python
# Display category to number mapping
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print("Category Mapping:")
print(mapping)
```

```python
print("TF-IDF Matrix Shape :", X_features.shape)
print("Number of Labels    :", len(y_encoded))
print("Number of Classes   :", len(label_encoder.classes_))
print("Classes             :", label_encoder.classes_)
```

```python
# ✅ Day 1 Verification Check
print("1. Rows in original clean_text vs feature matrix match:", X_features.shape[0] == len(y_encoded))
print("2. No missing values in labels:", pd.isnull(y_encoded).sum() == 0)
print("3. Number of classes matches expected categories:", len(label_encoder.classes_) == 6)
print("4. TF-IDF feature matrix is non-empty:", X_features.shape[1] > 0)
print("5. Sample check — first row has non-zero features:", X_features[0].nnz > 0)

print("\nSummary:")
print("TF-IDF Matrix Shape :", X_features.shape)
print("Number of Labels    :", len(y_encoded))
print("Number of Classes   :", len(label_encoder.classes_))
print("Classes             :", list(label_encoder.classes_))
```

```python
print("Original dataset rows:", len(df))
print("Rows used in feature engineering:", X_features.shape[0])
print("Difference:", len(df) - X_features.shape[0])
```

```python
import pandas as pd

df = pd.read_csv("../data/complaints_clean.csv")

print(df.shape)

df.head()
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
# Create TF-IDF Vectorizer

vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words="english",
    ngram_range=(1,2),
    min_df=2,
    max_df=0.95
)
```

```python
# Convert clean text into TF-IDF features

X_features = vectorizer.fit_transform(df["clean_text"])

print("TF-IDF Matrix Shape :", X_features.shape)
```

```python
feature_names = vectorizer.get_feature_names_out()

print("Total Features :", len(feature_names))
print()
print(feature_names[:20])
```

```python
import pickle

with open("../models/vectorizer.pkl", "wb") as file:
    pickle.dump(vectorizer, file)

print("✅ Vectorizer saved successfully!")
```

```python
import scipy.sparse

scipy.sparse.save_npz(
    "../models/tfidf_matrix.npz",
    X_features
)

print("✅ TF-IDF Matrix saved successfully!")
```

```python
import scipy.sparse
import pickle

# Load back what you saved
X = scipy.sparse.load_npz("../models/tfidf_matrix.npz")
with open("../models/vectorizer.pkl", "rb") as f:
    tfidf = pickle.load(f)

print("✅ Matrix shape:", X.shape)
print("✅ Vocabulary size:", len(tfidf.get_feature_names_out()))
print("✅ Sample features:", tfidf.get_feature_names_out()[:10])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_features,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_features,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)

print("Training Labels :", len(y_train))
print("Testing Labels :", len(y_test))
```

```python
model = MultinomialNB()

model.fit(
    X_train,
    y_train
)

print("Model trained successfully!")
```

```python
y_pred = model.predict(X_test)

print(y_pred[:10])
```

```python
accuracy = accuracy_score(
    y_test,
    y_pred
)

print("Accuracy :", accuracy)

print("\nClassification Report\n")

print(
    classification_report(
        y_test,
        y_pred
    )
)

print("\nConfusion Matrix\n")

print(
    confusion_matrix(
        y_test,
        y_pred
    )
)
```

```python
with open("../models/naive_bayes_model.pkl","wb") as file:
    pickle.dump(model,file)

print("Naive Bayes Model Saved!")
```

```python
results = pd.DataFrame({

    "Actual":y_test,

    "Predicted":y_pred

})

results.to_csv(

    "../models/day3_predictions.csv",

    index=False

)

print("Predictions Saved!")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

log_model.fit(X_train, y_train)

print("✅ Logistic Regression Model Trained Successfully!")
```

```python
log_pred = log_model.predict(X_test)

print(log_pred[:10])
```

```python
log_accuracy = accuracy_score(y_test, log_pred)

print("Logistic Regression Accuracy :", log_accuracy)

print("\nClassification Report\n")
print(classification_report(y_test, log_pred))

print("\nConfusion Matrix\n")
print(confusion_matrix(y_test, log_pred))
```

```python
with open("../models/logistic_regression_model.pkl", "wb") as file:
    pickle.dump(log_model, file)

print("✅ Logistic Regression Model Saved!")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("✅ Random Forest Model Trained Successfully!")
```

```python
rf_pred = rf_model.predict(X_test)

print(rf_pred[:10])
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)

print("Random Forest Accuracy :", rf_accuracy)

print("\nClassification Report\n")
print(classification_report(y_test, rf_pred))

print("\nConfusion Matrix\n")
print(confusion_matrix(y_test, rf_pred))
```

```python
with open("../models/random_forest_model.pkl", "wb") as file:
    pickle.dump(rf_model, file)

print("✅ Random Forest Model Saved!")
```

```python
comparison = pd.DataFrame({
    "Model": [
        "Naive Bayes",
        "Logistic Regression",
        "Random Forest"
    ],
    "Accuracy": [
        accuracy,
        log_accuracy,
        rf_accuracy
    ]
})

print(comparison)

best_model = comparison.loc[comparison["Accuracy"].idxmax()]

print("\n🏆 Best Model:")
print(best_model)
```
