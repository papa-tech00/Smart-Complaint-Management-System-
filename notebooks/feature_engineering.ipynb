```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv("../data/complaints_clean.csv")

print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Input feature
X = df["clean_text"]

# Target label
y = df["category"]

print("Input (X):")
print(X.head())

print("\nTarget (y):")
print(y.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words="english"
)

# Convert text into numerical features
X_features = vectorizer.fit_transform(X)

# Print shape
print("TF-IDF Matrix Shape:", X_features.shape)
```

```python
print(vectorizer.get_feature_names_out()[:20])
```

```python
from sklearn.preprocessing import LabelEncoder

# Create Label Encoder
label_encoder = LabelEncoder()

# Convert categories into numbers
y_encoded = label_encoder.fit_transform(y)

# Print encoded labels
print("Encoded Labels:")
print(y_encoded[:10])
```

```python
# Display category to number mapping
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print("Category Mapping:")
print(mapping)
```

```python
print("TF-IDF Matrix Shape :", X_features.shape)
print("Number of Labels    :", len(y_encoded))
print("Number of Classes   :", len(label_encoder.classes_))
print("Classes             :", label_encoder.classes_)
```

```python
# ✅ Day 1 Verification Check
print("1. Rows in original clean_text vs feature matrix match:", X_features.shape[0] == len(y_encoded))
print("2. No missing values in labels:", pd.isnull(y_encoded).sum() == 0)
print("3. Number of classes matches expected categories:", len(label_encoder.classes_) == 6)
print("4. TF-IDF feature matrix is non-empty:", X_features.shape[1] > 0)
print("5. Sample check — first row has non-zero features:", X_features[0].nnz > 0)

print("\nSummary:")
print("TF-IDF Matrix Shape :", X_features.shape)
print("Number of Labels    :", len(y_encoded))
print("Number of Classes   :", len(label_encoder.classes_))
print("Classes             :", list(label_encoder.classes_))
```

```python
print("Original dataset rows:", len(df))
print("Rows used in feature engineering:", X_features.shape[0])
print("Difference:", len(df) - X_features.shape[0])
```

```python
import pandas as pd

df = pd.read_csv("../data/complaints_clean.csv")

print(df.shape)

df.head()
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
# Create TF-IDF Vectorizer

vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words="english",
    ngram_range=(1,2),
    min_df=2,
    max_df=0.95
)
```

```python
# Convert clean text into TF-IDF features

X_features = vectorizer.fit_transform(df["clean_text"])

print("TF-IDF Matrix Shape :", X_features.shape)
```

```python
feature_names = vectorizer.get_feature_names_out()

print("Total Features :", len(feature_names))
print()
print(feature_names[:20])
```

```python
import pickle

with open("../models/vectorizer.pkl", "wb") as file:
    pickle.dump(vectorizer, file)

print("✅ Vectorizer saved successfully!")
```

```python
import scipy.sparse

scipy.sparse.save_npz(
    "../models/tfidf_matrix.npz",
    X_features
)

print("✅ TF-IDF Matrix saved successfully!")
```
